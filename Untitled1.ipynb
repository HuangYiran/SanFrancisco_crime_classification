```r
data.terror.weap = data.terror[,c("weaptype1", "weaptype1_txt", "weaptype2", "weaptype2_txt", "weaptype3", "weaptype3_txt", "weaptype4", "weaptype4_txt")]
data.terror.weap
```

```r
# Topic1 恐袭都有哪些weapon type：11种 每种都分别介绍， 最好不要扯上地区
weap.type = unique(data.terror.weap[,c("weaptype1", "weaptype1_txt")])
weap.type
```

```r
# Topic 1 或者可以只选择用量最多的几个进行介绍
data.terror.weap_number = data.terror[, .(count = .N), .(weaptype1_txt)][order(count)]
data.terror.weap_number
```

```r
# Topic 2 每种weapon type使用的地区, 根据目标对进行搜索
data.terror.region = unique(data.terror[, c("region", "region_txt")])
data.terror.region
```

```r
get.max_item = function(data.terror) {
    data.terror = data.table(data.terror)
    max_count = max(data.terror$count)
    return(data.terror[count == max_count,])
}
```

```r
# 给出Topic 2中的最优配对
data.terror.region_and_weapon = data.terror[,c("region", "region_txt", "weaptype1", "weaptype1_txt")]
data.terror.region_and_weapon = data.terror.region_and_weapon[,.(count = .N), .(region_txt, weaptype1_txt)][order(region_txt,count)]
data.terror.region_and_weapon = ddply(data.terror.region_and_weapon, .(region_txt),get.max_item)
data.terror.region_and_weapon
```
