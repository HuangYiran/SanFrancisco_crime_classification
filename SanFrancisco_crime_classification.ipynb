```r
# 先检验数据的完整性，是否有空值
# 返回输入列中包含的NULL值的个数
num_NULL <- function(li) {
    tmp = li %in% NULL
    len = length(li[tmp])
    return(len)
}
```

```r
get_next_year = function(year) {
    tmp = as.numeric(format(year, "%Y"))
    tmp = tmp + 1
    tmp = as.character(tmp)
    tmp = paste(tmp, "01", "01", sep = "-")
    next_year = as.Date(tmp)
    return(next_year)
}
```

```r
# 可以看出时间是从2003年的一月到2015年的的五月，最简单的先看一下每年的犯罪数量的变化
get_number_of_crime_in_year = function(year, data) {
    # 取data数据集中对应年份的子集，并计算当年的犯罪总数量
    # year是date类型数据
    this_year = format(year, "%Y")
    this_year = paste(this_year, "01", "01", sep = "-")
    this_year = as.Date(this_year)
    print(this_year)
    next_year = get_next_year(this_year)
    print(next_year)
    data = subset(data, data$Times > this_year & data$Times < next_year)
    data$Count = length(data$Dates)
    return(data)
}
```
